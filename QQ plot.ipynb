```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to generate and plot Q-Q plots
def generate_qq_plot(data, title):
    plt.figure(figsize=(6, 6))
    stats.probplot(data, dist="norm", plot=plt)
    plt.title(title)
    plt.show()

# 1. Normally Distributed Data
normal_data = np.random.normal(0, 1, 1000)

# 2. Left-Skewed Data (using a negative skewness)
left_skewed_data = np.random.chisquare(2, 1000) * -1

# 3. Right-Skewed Data (using a positive skewness)
right_skewed_data = np.random.chisquare(2, 1000)

# Plotting Q-Q Plots
generate_qq_plot(normal_data, "Q-Q Plot: Normally Distributed Data")
generate_qq_plot(left_skewed_data, "Q-Q Plot: Left-Skewed Data")
generate_qq_plot(right_skewed_data, "Q-Q Plot: Right-Skewed Data")
```

```python
# 4. Approximately Normally Distributed Data (with slight random noise)
approx_normal_data = np.random.normal(0, 1, 1000) + np.random.uniform(10, 20, 1000)

# Plotting the Q-Q Plot for Approximately Normally Distributed Data
generate_qq_plot(approx_normal_data, "Q-Q Plot: Approximately Normally Distributed Data")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range for x-axis
x = np.linspace(-5, 15, 400)

# Calculate the normal distribution curves
y1 = (1 / np.sqrt(2 * np.pi * 1)) * np.exp(-0.5 * ((x - 2) ** 2) / 1)
y2 = (1 / np.sqrt(2 * np.pi * 2)) * np.exp(-0.5 * ((x - 5) ** 2) / 2)
y3 = (1 / np.sqrt(2 * np.pi * 0.5)) * np.exp(-0.5 * ((x - 3) ** 2) / 0.5)
y4 = (1 / np.sqrt(2 * np.pi * 1)) * np.exp(-0.5 * ((x - 10) ** 2) / 1)
print(y2[:10])
# Plot the curves
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label="Mean = 2, Variance = 1")
plt.plot(x, y2, label="Mean = 5, Variance = 2")
plt.plot(x, y3, label="Mean = 3, Variance = 0.5")
plt.plot(x, y4, label="Mean = 10, Variance = 1")

# Labeling the mean and variance
plt.axvline(x=2, color='blue', linestyle='--', ymax=0.5)
plt.axvline(x=5, color='orange', linestyle='--', ymax=0.5)
plt.axvline(x=3, color='green', linestyle='--', ymax=0.5)
plt.axvline(x=10, color='red', linestyle='--', ymax=0.5)

# Add labels and title
plt.title("Normal Distribution Curves with Different Means and Variances")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np

# Define the range for x-axis (10 data points)
x_points = np.linspace(0, 10, 10)

# Define means and variances
mu1, sigma1_sq = 2, 1
mu2, sigma2_sq = 5, 2
mu3, sigma3_sq = 3, 0.5
mu4, sigma4_sq = 10, 1

# Compute the y2 values for the 10 data points
y2_values = (1 / np.sqrt(2 * np.pi * sigma2_sq)) * np.exp(-0.5 * ((x_points - mu2) ** 2) / sigma2_sq)

# Function to compute likelihood
def compute_likelihood(y_values, mu, sigma_sq):
    likelihood = (1 / np.sqrt(2 * np.pi * sigma_sq)) * np.exp(-0.5 * ((y_values - mu) ** 2) / sigma_sq)
    return np.sum(np.log(likelihood))

# Calculate likelihoods for y2 data points under other distributions
likelihood_y1 = compute_likelihood(y2_values, mu1, sigma1_sq)
likelihood_y2 = compute_likelihood(y2_values, mu2, sigma2_sq)
likelihood_y3 = compute_likelihood(y2_values, mu3, sigma3_sq)
likelihood_y4 = compute_likelihood(y2_values, mu4, sigma4_sq)

likelihood_y1, likelihood_y2, likelihood_y3, likelihood_y4
```

```python
# Determine the maximum log-likelihood
max_log_likelihood = max(likelihood_y1, likelihood_y2, likelihood_y3, likelihood_y4)
max_log_likelihood
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Set seed for reproducibility
np.random.seed(42)

# Parameters for the synthetic time series
n = 100  # Number of data points
phi = 0.8  # AR(1) coefficient for inducing autocorrelation
c = 2  # Constant term

# Generate synthetic time series with autocorrelated residuals
y = np.zeros(n)
epsilon = np.random.normal(0, 1, n)
for t in range(1, n):
    y[t] = c + phi * y[t-1] + epsilon[t]

# Fit a simple linear model to the data
X = sm.add_constant(np.arange(n))  # Adding a constant for the intercept
model = sm.OLS(y, X)
results = model.fit()

# Get the residuals
residuals = results.resid

# Plot the residuals
plt.figure(figsize=(10, 6))
plt.plot(residuals, label='Residuals', color='blue')
plt.title('Residuals of the Synthetic Time Series')
plt.xlabel('Time')
plt.ylabel('Residuals')
plt.axhline(y=0, color='black', linestyle='--')
plt.legend()
plt.show()

# Show the autocorrelation in residuals
sm.graphics.tsa.plot_acf(residuals, lags=30)
plt.show()
```

```python
# Generate a synthetic time series with heteroskedastic residuals
n = 100  # Number of data points
x = np.linspace(0, 10, n)
y = 2 + 0.5 * x + np.random.normal(0, 1, n) * (1 + 0.5 * x)  # The noise increases with x

# Fit a simple linear model to the data
X = sm.add_constant(x)  # Adding a constant for the intercept
model = sm.OLS(y, X)
results = model.fit()

# Get the residuals
residuals = results.resid

# Plot the residuals
plt.figure(figsize=(10, 6))
plt.scatter(x, residuals, label='Residuals', color='blue')
plt.title('Residuals of the Synthetic Time Series (Heteroskedastic)')
plt.xlabel('Time')
plt.ylabel('Residuals')
plt.axhline(y=0, color='black', linestyle='--')
plt.legend()
plt.show()

# Plot the squared residuals to highlight heteroskedasticity
plt.figure(figsize=(10, 6))
plt.scatter(x, residuals**2, label='Squared Residuals', color='red')
plt.title('Squared Residuals of the Synthetic Time Series (Heteroskedastic)')
plt.xlabel('Time')
plt.ylabel('Squared Residuals')
plt.axhline(y=0, color='black', linestyle='--')
plt.legend()
plt.show()
```
